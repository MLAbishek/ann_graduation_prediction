```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df=pd.read_csv(r"D:\programming\models\dataset\train_grad.csv")
df.head()
```

```python
np.unique(df['Target'])
```

```python
df.select_dtypes('object')
```

```python
from sklearn.preprocessing import LabelEncoder
lr=LabelEncoder()
df['Target']=lr.fit_transform(df['Target'])
lr.classes_
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
print(x.shape,y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
print(x)
```

```python
ann=tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=64,activation='relu'))
ann.add(tf.keras.layers.Dense(units=128,activation='relu'))
ann.add(tf.keras.layers.Dense(units=256,activation='relu'))
ann.add(tf.keras.layers.Dense(units=3,activation='softmax'))
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=[['accuracy']])
ann.fit(x,y,epochs=250,batch_size=32)
```

```python
test_df=pd.read_csv(r"D:\programming\models\dataset\test_grad.csv",index_col=False)
```

```python
test_df.isnull().sum()
```

```python

val=test_df.iloc[:,1:].values
val=sc.transform(val)
```

```python
y_pred=ann.predict(val)
print(y_pred)
```

```python
y_pred=np.argmax(y_pred,axis=1)
y_pred
```

```python
y_pred=pd.DataFrame(y_pred)
print(y_pred)
```

```python
y_pred=y_pred.replace(0,'Dropout')
y_pred=y_pred.replace(1,'Enrolled')
y_pred=y_pred.replace(2,'Graduate')
print(y_pred)
```

```python
frame = [test_df[['id']], y_pred]
ans_df = pd.concat(frame, axis=1) 
```

```python
ans_df.to_csv("grad_ans.csv",index=False)
```
